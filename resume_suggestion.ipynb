```python
with open("resume.txt","r") as f:
    resume=f.read()
```

```python
print(resume)
```

```python
extraction_prompt= f"""You are a professional resume counseler. Based on the following resume, please extract the key skills, experiences, and qualifications that are relevant to a job application. Please provide a summary of the candidate's strengths and areas of expertise.
resume: {resume}"""
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
extracted_info=llm.invoke(extraction_prompt)
extracted_info=extracted_info.content
```

```python
extracted_info
```

```python
designation_prompt = f"""Based on the extracted information from the resume, please suggest potential job roles or job designations that would be good fit for the candidate. Consider the candidate's skills, experiences, and qualifications when making your suggestion. Please provide a list of potential job titles with a brief explanation of why each role would be suitable for the candidate.
Extracted Information: {extracted_info}"""
```

```python
roles_suggested=llm.invoke(designation_prompt)
```

```python
print(roles_suggested.content)
```

```python
import base64
```

```python
with open("resume.png","rb") as f:
    image_encoded=base64.b64encode(f.read()).decode()
```

```python
image_encoded
```

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model="meta-llama/llama-4-scout-17b-16e-instruct")
```

```python
from langchain_core.messages import HumanMessage
human_message=HumanMessage(content=[
    {"type": "image_url", "image_url": {"url": f"data:image/png;base64,{image_encoded}"}},
    {"type": "text",      "text": "Extract all the resume fields, languages,projects and tools the the file. dont make any errors"},
])
```

```python
response=llm.invoke([human_message])
```

```python
print(response.content)
```

```python
response.formatted=response.content
```

```python
designation_prompt2 = f"""Based on the extracted information from the resume, please suggest potential job roles or job designations that would be good fit for the candidate. Consider the candidate's skills, experiences, and qualifications when making your suggestion. Please provide a list of potential job titles with a brief explanation of why each role would be suitable for the candidate.
Extracted Information: {response.formatted}"""
```

```python
roles_suggested=llm.invoke(designation_prompt2)
```

```python
print(roles_suggested.content)
```
